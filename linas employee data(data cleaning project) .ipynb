```python
## import relevant libraries 
import pandas as pd 
import numpy as np
```

```python
## import file 
data = pd.read_csv(r"C:\Users\elyon\OneDrive\Documents\GitHub\Linas Employee data.csv")
```

```python
data.head(50)
```

```python
# data cleaning will include :
# correct datatype
# Handle null values 
# handle NaN values 
# Round up figures 
# normalize data
# merge columns 
```

```python
## check for duplicates 
data.duplicated().sum()
```

```python
data.describe()
```

```python
data.info()
```

```python
#Replace NaN With the Mean Value
data['salary'].fillna(data['salary'].mean(), inplace = True)
data.head(9)
```

```python
# rename columns
data = data.rename(columns = {'employee_id' : 'Employee_Id', 'first_name' : 'First_name', 'last_name':'Last_name', 'department' : 'Department', 'hire_date' : 'Hire_date', 'performance_rating' : 'Performance_rating', 'salary' : 'Salary', 'tenure_years' : 'Tenure-years', 'attendance_percentage' : 'Attendance_percentage'})
data
```

```python
# join first and last name columns using the  concat function 
data['Employee_name'] = data['First_name'] + ' ' +  data['Last_name']
```

```python
data.head()
```

```python
## drop irrelevant columns
data.drop(['First_name', 'Last_name'], axis = 1)
```

```python
##reorder columns 
data = data[['Employee_Id', 'Employee_name', 'Department', 'Hire_date', 'Performance_rating', 'Salary', 'Attendance_percentage', 'Level', 'Tenure-years']]
```

```python
data.head()
```

```python
data.rename(columns = {'Tenure-years' : 'Tenure_years'})
```

```python
data.loc[:, 'Hire_date'] = pd.to_datetime(data.loc[:, 'Hire_date'])
```

```python
data.head()
```

```python
## normalize data 
data.loc[:, 'Level'] = data['Level'].str.lower()
```

```python
data.loc[:,'Department'] = data['Department'].str.lower()
```

```python
data.loc[:, 'Employee_name'] = data['Employee_name'].str.lower()
```

```python
data.head(20)
```

```python
data.loc[:, ['Salary']] = data[['Salary']].apply(np.ceil)
```

```python
data.head()
```

```python
## export data 
data.to_csv('Lina_s employee data.csv', index = False)
```
